#### **Variables**

- A variable refers to a memory address in which data is stored. Number at the beginning, special character, hyphen are not allowed when naming a variable. A variable can have a short name (like x, y, z), but a more descriptive name (firstname, lastname, age, country) is highly recommended.

* Python Variable Name Rules

 - A variable name must start with a letter or the underscore character
 - A variable name cannot start with a number
 - A variable name can only contain alpha-numeric characters and underscores (A-z, 0-9, and _ )
 - Variable names are case-sensitive (firstname, Firstname, FirstName and FIRSTNAME) are different variables)

```python
# here are some example of valid variable name 
name = 'Kapil'
last_name = 'saran'
First_Name = 'kya karo ge jaan ke'
_What_is_your_name = 'kyo'
age = 23
num_1 = 19
num_2 = 20
year_2021 = 2025
_if = 'agar'
```

- Let us use the print() and len() built-in functions. Print function takes unlimited number of arguments. An argument is a value which we can be passed or put inside the function parenthesis, see the example below.

```python
print('kapil','Saran')
print('kapil',3)
print('kapil'*3)
print('kapil',',','saran')
print(len('kapil saran'))
```

```python
print('kapil saran',end='')
print('kapil saran',sep = '*')
```

```python
print('kapil saran', end='\n')
print('kapil saran', sep='*')
```

- end='\n' tells print() to end with a newline (this is the default behavior).
- sep='*' specifies the separator between multiple arguments passed to print(). Since you're only printing one argument ('kapil saran'), sep has no effect.

**Casting:** Converting one data type to another data type. We use int(), float(), str(), list, set When we do arithmetic operations string numbers should be first converted to int or float otherwise it will return an error. If we concatenate a number with a string, the number should be first converted to a string. We will talk about concatenation in String section.

```python
# int to float
num_int = 10
print('num_int',num_int)         # 10
num_float = float(num_int)
print('num_float:', num_float)   # 10.0

# float to int
gravity = 9.81
print(int(gravity))             # 9

# int to str
num_int = 10
print(num_int)                  # 10
num_str = str(num_int)
print(num_str)                  # '10'

# str to int or float
num_str = '10.6'
num_float = float(num_str)  # Convert the string to a float first
num_int = int(num_float)    # Then convert the float to an integer
print('num_float', float(num_str))  # 10.6
num_int = int(num_float)
print('num_int', int(num_int))      # 10

# str to list
first_name = 'Asabeneh'
print(first_name)               # 'Asabeneh'
first_name_to_list = list(first_name)
print(first_name_to_list)            # ['A', 's', 'a', 'b', 'e', 'n', 'e', 'h']
```
